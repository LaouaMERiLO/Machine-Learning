**Types of AI:**

There are four types of artificial intelligence: reactive machines, limited memory, theory of mind and self-awareness.


1. **Reactive machines:**
  * The most basic types of AI systems are purely reactive, and have the ability neither to form memories nor to use past experiences to inform current decisions.
  * This type of intelligence involves the computer perceiving the world directly and acting on what it sees. It doesn’t rely on an internal concept of the world.
  * In a seminal paper, AI researcher Rodney Brooks argued that we should only build machines like this.
  * They can’t interactively participate in the world, the way we imagine AI systems one day might.
  * Instead, these machines will behave exactly the same way every time they encounter the same situation. This can be very good for ensuring an AI system is trustworthy: You want your autonomous car to be a reliable driver.
  * But it’s bad if we want machines to truly engage with, and respond to, the world. These simplest AI systems won’t ever be bored, or interested, or sad.
2. **Limited memory:**
  * Machines in this class can look into the past.
  * Characterized by the ability to absorb learning data and improve over time similar to the way of humans.
  * This is the AI that is widely used and being perfected today (Deep learning algorithms).
  * Self-driving cars observe other cars’ speed and direction over time. This infomration will lateron be used to take decisions such as turning or speeding up or down.
  * These simple pieces of information about the past are only transient. They aren’t saved as part of the car’s library of experience it can learn from, the way human drivers compile experience over years behind the wheel.
  * So, is it possible to put forward an AI system that builds full representations, remember experiences and learn how to handle new situations?
3. **Theory of mind:**
  * It is here we distinguish between the machines we have nowdays and the ones we will build in the future.
  * It is important to discuss the types of representations such machines need to form, and what they need to be about.
  * It is not sufficient to form representations about the world only, but also about other agents or entities in the world.
  * In psychology, this is called *“theory of mind”* – understanding that people, creatures and objects in the world have thoughts and emotions that affect their own behavior.
  * This is crucial to how we humans formed societies, because they allowed us to have social interactions.
4. **Self-awareness:** (Singularity)
  * The final step of AI development is to build systems that can form representations about themselves.
  * Ultimately, we will have to not only understand consciousness, but build machines that have it.

  * algorithms are trained using labeled examples (inputs where the desired outputs are known).

2. **Unsupervised learning:**
  * Used against data that has no historical labels (The system is not told the "right answer.")

  * used for the same applications as supervised learning.

4. **Reinforcement learning:**
  * Often used for robotics, gaming and navigation.
  * The algorithm doesn't need historical samples, but discovers through trial and error which actions yield the greatest rewards.
  * This type of learning has three primary components: the agent, the environment and actions. The objective is for the agent to choose actions that maximize the expected reward over a given amount of time.

* The global performance of the model can be assessed using the **Cost** function defined as:

  $J(θ) = \frac{∑_{i=1}^{n}{L(h_θ(x_i), y_i)}}{n}$

* To reduce cost, the configuration of the model $θ_i$ needs to be repalced with another configuration $ θ_{i+1}$. This procedure is called **Gradient Descent**

  $θ_{i+1} = θ_i -ŋ ∇ J(θ_i)$, where $ŋ$ is the learning rate (the step scaling factor).

* In general, the likelihood of a model $Y(\theta)$ (scoring function) given parameters $θ$ is used to find the optimal parameters $\theta$ through likelihood maximization:

  $θ^{opt} = argmax(Y(\theta)) $

* in practice, the used log-likelihood is $\log(Y(\theta))$ which is easier to optimize.

**Example:**
Let"s supose that we have the following model (function) for which we want to find the optimale parameters.

$x^4 + 2.5x^3 - 0.5x^2 -x + 1$



```python
import matplotlib.pyplot as plt
import numpy as np

def model(x):
  return x**4 + 2.5*x**3 - 0.5*x**2 -x + 10

plt.plot( np.arange(-3,2,0.01), model(np.array(np.arange(-3,2,0.01))));
```

The derivative of the model to be used to calculate the gradient at each point

```python
def deriv_model(x):
  return 4*x**3 + 7.5*x**2 - x - 1

plt.plot(np.arange(-3,2,0.01), model(np.array(np.arange(-3,2,0.01))));
plt.plot(np.arange(-3,2,0.01), deriv_model(np.array(np.arange(-3,2,0.01))));
plt.axhline(y=0, color='r', linestyle='-')
```

an implementation of the Gradient Descent algorithm to be used.

```python
def gradient_descent(start, gradient, learn_rate, max_iter, tol=0.01):
  # 1. starting point - often a random initialisation
  # 2. gradient function - has to be specified before-hand
  # 3. learning rate - scaling factor
  # 4. maximum number of iterations
  # 5. tolerance to conditionally stop the algorithm (in this case a default value is 0.01)

  steps = [start] # history tracking
  x = start

  for _ in range(max_iter):
    diff = learn_rate*gradient(x)
    if np.abs(diff)<tol:
      break
    x = x - diff
    steps.append(x) # history tracing

  return steps, x
```

* **Least Mean Squares (LMS) algorithm:**
  * Somtimes, we have fewer computational resources and memory.
  * In such cases, LMS algorithms may be used since they do not involve any matrix operations.
  * LMS algorithm is a special case of gradient descent algorithm.
  * By noting $α$ the learning rate, the update rule of the Least Mean Squares (LMS) algorithm for a training set of $m$ data points is as follows:

      $\forall j, \theta_{j+1} \longleftarrow \theta_j + \alpha \sum_{i=1}^{m}\left[ y_i - h_\theta(x_i) \right]x_i$

**Example:**

The following code presents an example of how LMS works.


1.   Impliment the model/hypothesis $h_θ$



```python
def hθ(x, θ):
    return θ.dot(x)
```

2. Impliment the cost function $L$

```python
def Cost(y, x, θ, factor = 1):
  return (np.power(hθ(x, θ) - y, 2) * factor).sum()
```

3. The code for the LMS function

```python
def LMS(x, y, α, θ):
  Cmax = 1e+20
  maxIter = 200
  i = 0
  while ((i < maxIter) and (Cost(y, x, θ) < Cmax)):
    # print(Cost(y, x, θ))
    θ = θ + α * (Y - hθ(X, θ)).dot(X.transpose())
    i = i +1
  return θ
```

```python
#set inititla value for θ

α = 5e-5
# An example of data X
X = np.array([[1, 2], [4, 6], [5, 12], [4, -14], [-3, 0],
              [0, 0], [5, 12], [-3, -14], [1, 14], [5, 5]]).transpose()
# data Y
Y = np.array([.3, .10, .128, .27, -.413, 0, .17, -.17, .1005, .102]).transpose()

# Loop with random initital hypotheses
for x in range(25):
  θ = (1 - np.random.random(2)) * 4
  d = LMS(X, Y, α, θ)
  print('begening cost:', Cost(Y, X, θ), θ, ', the final cost is : ', Cost(Y, X, d), np.round(d, 2))
```

**Softmax regression:**
* A softmax regression, also called a multiclass logistic regression.
* It is used to generalize logistic regression when there are more than 2 outcome classes.
* the Bernoulli hypothesis $h_{θi}$  of each class is:

<center>$ h_{θ_i}(X) = Pr(Y=i∣X;θ_i) = \frac{e^{θ_i^TX}}{∑_{i=1}^{m} e^{θ_i^TX}}$, where $m$ is the number of classes. </center>

Example:

```python
import numpy as np

class SoftmaxRegression(LogisticRegression):
    def __init__(self, learning_rate=0.1, num_iterations=1000):
        super().__init__(learning_rate=learning_rate, num_iterations=num_iterations)

    def fit(self, X, y, verbos=False):
        # initialize weights and bias to zeros
        self.weights = np.zeros((X.shape[1], len(np.unique(y))))

        # one-hot encode labels
        y_encoded = np.zeros((len(y), len(np.unique(y))))
        for i, label in enumerate(y):
            y_encoded[i][label] = 1


        # update weights iteratively
        for i in range(self.num_iterations):
            z = np.dot(X, self.weights)
            y_pred = self.softmax(z)
            dw = np.dot(X.T, (y_pred - y_encoded)) / len(y)
            self.Costs.append(np.sum((y_pred - y_encoded)**2) / len(y))
            if verbos:
                print('Current cost (MSE) :', np.sum((y_pred - y_encoded) ** 2) / len(y))
            self.weights -= self.learning_rate * dw
        # self.Costs = np.reshape(self.Costs, (-1,3))
    def predict(self, X):
        z = np.dot(X, self.weights)
        y_pred = self.softmax(z)
        return np.argmax(y_pred, axis=1)

    def softmax(self, z):
        return np.exp(z) / np.sum(np.exp(z), axis=1, keepdims=True)
```

```python
import numpy as np

def predictNbrPeopleLinearModel(y, temp):
  return y  + (temp * 100)

print ('temp -20 (Linear) : ', predictNbrPeopleLinearModel(5000, -20))
print ('temp -10 (Linear) : ', predictNbrPeopleLinearModel(50, -10))

def predictNbrPeopleExpModel(y, temp):
  return y +  np.sign(temp)*((y/2)*(pow(1/2, 1 + abs(abs(temp) - 10)/10)-1)/ (1/2 - 1))

print ('temp -20 (Exp) : ', predictNbrPeopleExpModel(5000, -20))
print ('temp -10 (Exp) : ', predictNbrPeopleExpModel(50, -10))
```

#Exponential family
* The exponential family is a parametric set of probability distributions with useful algebraic properties.
* A single-parameter exponential family has a probability density function of the form: $${\displaystyle f(x\mid \theta )=h(x)\exp {\bigl [}\eta (\theta )\cdot T(x)-A(\theta ),{\bigr ]}}$$
* The natural parameter or canonical parameter of the distribution is represented by $\eta(\theta)$ and maps the parameter of interest $\theta$ to the space in which the sufficient statistic $T(x)$ lives.
* The carrier density or base measure of the distribution is represented by $h(x)$ and ensures the probability density function is non-negative.
* The log-partition function or cumulant generating function of the distribution is represented by $A(\theta)$ and ensures the total probability of the distribution sums to 1, given by: $${ A(\theta )=\log \int _{X}h(x)\,\exp(\eta (\theta )\cdot T(x))\,\mathrm {d\,} x}$$
* The sufficient statistic $T(x)$ is a function of the observed data $x$ that contains all relevant information about the parameter of interest $\theta$.
* Once we know the value of $T(x)$, the rest of the data is redundant and doesn't provide additional information about $\theta$.
* The exponential family includes many commonly used distributions, such as the normal, Poisson, and exponential distributions, as special cases.
* The exponential family is a flexible and powerful framework for statistical modeling.

**Assumptions of GLMs:**

Generalized Linear Models (GLM) aim at predicting a random variable y as a function of $ x\in\mathbb{R}^{n+1}$  and rely on the following 3 assumptions:

* **$y|x, θ \sim ExpFamily(η) $**:
  * The data  are independently distributed
  * The dependent variable  assumes a distribution from an exponential family (e.g. binomial, Poisson, multinomial, normal, etc.).
* **$h_θ(x) = E[x|y,θ]$**:

  * $E$ is the conditional expectation of $x$ given $y$ under the configuration $θ$.
* **$η = θ^{T} X$**:
  * GLM does NOT assume a linear relationship between the response variable and the explanatory variables, but it does assume a linear relationship between the transformed expected response in terms of the link function ($\eta$) and the explanatory variables;
  * For example, in logistic regression, the link function is the logit function, which transforms the probability of the response variable being in a certain category to the log odds of being in that category. This allows us to model the relationship between the explanatory variables and the log odds of the response variable being in a certain category, in a linear way.

ordinary linear and logistic regressions are special cases of generalized linear models.

**Example:** Suppose you want to predict the probability of a student passing a test based on their study hours and which group they attend (G1 or G2). GLM could be used to model this problem using a binomial distribution and a logit link function. The model could then predict the probability of passing the test based on the student's paased time in the exam and attendance to a specefic study group.

```python
import pandas as pd
import statsmodels.api as sm

# load data into a Pandas DataFrame
url="https://raw.githubusercontent.com/mafudge/datasets/master/exam-scores/exam-scores.csv"
data = pd.read_csv(url)


# fit a GLM with a binomial distribution and logit link function
data['passed'] = data['Student_Score'] > 15
data['attended_group'] = (data['Class_Section'] == 'M01').astype(int)

model = sm.GLM(data[['passed']], data[['Completion_Time', 'attended_group']], family=sm.families.Binomial())
result = model.fit()

# predict probability of passing test for a new student with 5 hours of studying and didn't attend a study group
new_student = [[25, 1]]
prob_pass = result.predict(new_student, linear=False)[0]

print("The predicted probability of passing the test for a student who spent 30 mins and did attend group M01 is:", np.round(prob_pass*100, 2), '%')
```

# Support Vector Machines

* Support Vector Machine (SVM) is a supervised machine learning algorithm that can be used for both classification or regression challenges.
* The goal of support vector machines is to find the linev(hyperplane) that maximizes the minimum distance to the line (hyperplane).
* Support Vectors are simply the coordinates of individual points which suport the line (hyperplane).
* It uses a subset of training points in the decision function, so it is also memory efficient.
<center>
<img src="https://upload.wikimedia.org/wikipedia/commons/thumb/7/72/SVM_margin.png/450px-SVM_margin.png" width=400></img><center>

**How it works**
* supose that we have the data ${ (\mathbf {x} _{1},y_{1}),\ldots ,(\mathbf {x} _{n},y_{n}),}$, where $y_i ∈ \{1, -1\}$
* The goal is to find the "maximum-margin hyperplane" that divides the group of points ${x} _{i}$ for which $y_{i}=1$ from the group of points for which $y_{i}=-1$.

    <center>${\mathbf {w} ^{T}\mathbf {x} -b=0}$</center>
* $ {w} $ is the normal vector to the hyperplane. $\tfrac {b} {||\mathbf{w}||}$ determines the offset of the hyperplane from the origin along the normal vector.
* The distance between these two suport lines (hyperplanes) is ${\tfrac {2}{\|\mathbf {w} \|}}$.
* So, the problem is how to maximize the distance between the lines such that
<center>
${ \mathbf {w} ^{T}\mathbf {x} _{i}-b\geq 1}, if  { y_{i}=1}$ 𝝠
${ \mathbf {w} ^{T}\mathbf {x} _{i}-b\leq -1}, if { y_{i}=-1}$.
</center>
or :
<center>
$Min||\mathbf {w}||^{2}$, subject to $y_i(\mathbf {w}^Tx_i-b) ≥ 1$
</center>
* The $\mathbf {w}$  and $b$ that solve this problem determine our classifier, $ \mathbf {x} \mapsto \operatorname {sign}(\mathbf {w} ^{T}\mathbf {x} -b)$



**How to optimize:**
* The ***hinge loss*** is used in the setting of SVMs and is defined as follows:
$L(z,y) = max (0, 1 - yz) = max (0, 1 - y×({w} ^{T}\mathbf {x} -b))$
* The **cost** is therefore:

  $j(w) = \frac{λ}{2} ||w||^2 + \frac{1}{N} \sum_{i}^{N} L(z_i,y_i)$
* The **gradient** of the cost function will be:

   $Δ_w j(w) = λw - \frac{1}{M} \sum_{i}^{M} y_ix_i$

The following code implements a basic Support Vector Machine (SVM) classifier using the hinge loss function and gradient descent optimization.

```python
class SVM:
    def __init__(self, learning_rate=0.001, regularization_param=0.1, num_iterations=1000):
        self.learning_rate = learning_rate
        self.regularization_param = regularization_param
        self.num_iterations = num_iterations
        self.w = None
        self.b = None

    def fit(self, X, y):
        n_samples, n_features = X.shape
        y_ = np.where(y <= 0, -1, 1)  # convert y values to -1 and 1
        self.w = np.zeros(n_features)  # initialize weight vector to zeros
        self.b = 0  # initialize bias term to zero

        # loop for the given number of iterations
        for _ in range(self.num_iterations):
            # loop over all training samples
            for idx, x_i in enumerate(X):
                # calculate condition for the hinge loss function
                condition = y_[idx] * (np.dot(x_i, self.w) - self.b) >= 1

                # update weight and bias based on condition
                if condition:
                    self.w -= self.learning_rate * (2 * self.regularization_param * self.w)
                else:
                    self.w -= self.learning_rate * (2 * self.regularization_param * self.w - np.dot(x_i, y_[idx]))
                    self.b -= self.learning_rate * y_[idx]

    def predict(self, X, is_approx = False):
        # calculate approximate values for the prediction
        approx = np.dot(X, self.w) - self.b
        # return the sign of the approximate values
        if is_approx:
          return approx
        return np.sign(approx)
```

```python
import numpy as np
from sklearn import datasets
import matplotlib.pyplot as plt

# Load iris dataset and extract first two classes
iris = datasets.load_iris()
X = iris.data[iris.target != 2, :2]  # extract first two features
y = iris.target[iris.target != 2]
y[y == 0] = -1  # convert labels to -1 and 1

# Initialize SVM model and fit to data
svm = SVM( learning_rate=0.001, regularization_param=0.005, num_iterations=1000)
svm.fit(X, y)

# Plot decision boundary and support vectors
plt.figure(figsize=(16,8))
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Paired)
xlim = plt.xlim()
ylim = plt.ylim()
xx, yy = np.meshgrid(np.linspace(xlim[0], xlim[1], 100), np.linspace(ylim[0], ylim[1], 100))
Z = svm.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.contour(xx, yy, Z)
plt.scatter(X[np.abs(np.dot(X, svm.w) - svm.b) <= 1, 0], X[np.abs(np.dot(X, svm.w) - svm.b) <= 1, 1], s=100, marker='o', facecolors='none', edgecolors='k')
plt.xlim()
plt.ylim()
plt.show()

# Calculate accuracy on training set
y_pred = svm.predict(X)
accuracy = np.mean(y_pred == y) * 100
print(f"Accuracy on training set: {accuracy:.2f}%")
```

## Soft margin SVM:
*  The Soft Margin SVM extends the standard SVM algorithm and adds a slack variable to the original SVM formulation.
* It allows for some misclassifications in the training data, which is useful when the data is not perfectly separable or there are outliers or noise.
* The objective function of the Soft Margin SVM minimizes the sum of the regularization term and the hinge loss function.
* The Soft Margin SVM can handle non-linearly separable datasets and is more robust to noisy or unbalanced data. The hyperparameter C controls the trade-off between the margin size and the number of misclassifications.

* A new varaiable $ξ$, called slack variable, to decide whetether a data ponit is linearly separable or not.
* The regularization term penalizes the slack variables and helps to balance the trade-off between maximizing the margin and minimizing the misclassification errors.
  * **if $ξ_i$= 0**, the points can be considered as correctly classified.
  * **else** , Incorrectly classified points.
* The previous problem can now be rewriten it as:

<center>

## The Kernel trick
* Somtimes, the data set is not at all linearly separable.
* The kernel trick converts non-linear lower dimension space to a higher dimension space thereby we can get a linear classification.
<center><img width=500 src="https://miro.medium.com/max/838/1*gXvhD4IomaC9Jb37tzDUVg.png"></img></center>
* Given a feature mapping $ϕ$, we define the kernel $K$ as follows:

<center>$K(x,z) = ϕ(x)^Tϕ(z)$</center>

* The types of kernel $K$ are: linear kernel,
polynomial kernel, Gaussian Kernel.
* In practice, the kernel $K$ defined by $K(x,z)=\exp\left(-\frac{||x-z||^2}{2\sigma^2}\right)$ is called the Gaussian kernel and is commonly used.
* The polynomial kernel is defined as: $K(x,z) = (a + x^Tz)^b$, where $b$ is degree of the kernel and $a$ is a constant term.

```python
import numpy as np
import matplotlib.pyplot as plt
import sklearn.datasets as ds
import seaborn as sns
from sklearn import svm

data = ds.make_circles(n_samples=200, factor=0.3, noise=0.08, random_state=0)
X = data[0]
Y = data[1]
# plt.scatter(X[:,0], X[:,1],  c=Y)

fig, ax = plt.subplots(nrows=2, ncols=3, figsize=(18, 5))
fignum = 0
for kernel in ("linear", "poly", "rbf"):
    clf = svm.SVC(kernel=kernel, gamma=2)
    clf.fit(X, Y)
    Z = clf.predict(X)
    ax[0,fignum].scatter(X[:, 0], X[:, 1], c=Z, zorder=10, cmap=plt.cm.Paired, edgecolors="k")
    fignum = fignum + 1


data = ds.make_moons(n_samples=100, noise=0.05)
X = data[0]
Y = data[1]

fignum = 0
for kernel in ("linear", "poly", "rbf"):
    clf = svm.SVC(kernel=kernel, gamma=3, degree = 3)
    clf.fit(X, Y)
    Z = clf.predict(X)
    ax[1,fignum].scatter(X[:, 0], X[:, 1], c=Z, zorder=10, cmap=plt.cm.Paired, edgecolors="k")
    fignum = fignum + 1

```

Multi-class SVM:

* Multi-class SVM is an extension of binary SVM to the multi-class classification problem.
* In binary classification, the SVM learns a decision boundary to separate the data points of two classes. However, in multi-class classification, we have more than two classes, so we need to learn multiple decision boundaries to separate the data points of different classes.
* There are two main approaches for multi-class SVM: One-vs-One (OVO) and One-vs-All (OVA).
<center>
<img src ='https://ars.els-cdn.com/content/image/3-s2.0-B9780123849885000206-f20-02-9780123849885.jpg' />
</center>
 * In OVO, a separate binary classifier is trained for each pair of classes. For example, if we have three classes (A, B, and C), we will train three binary classifiers (AB, AC, and BC). During testing, we apply each classifier to the test data and count the number of votes for each class to make the final decision.
 * In OVA, we train a single binary classifier for each class, which learns to distinguish that class from all other classes. During testing, we apply all the binary classifiers to the test data and assign the test point to the class with the highest score.
* The OVO approach is generally used when the number of classes is small because it requires training many binary classifiers, while OVA is used when the number of classes is large because it requires training only one binary classifier for each class.
* Multi-class SVM can also be extended to non-linearly separable data using kernel functions in the same way as binary SVM.
* The Soft Margin SVM and kernel trick concepts can be extended to multi-class SVM, allowing us to handle more complex data and achieve better performance.


# Generative Learning

* A generative model first tries to learn how the data is generated by estimating $P(x|y)$.
*  The prior probability $P(Y)$ and the likelihood probability $P(X|Y)$ are estimated with the help of the training data.
* Finaly, ***Bayes Theorem*** is employed to calculate the posterior probability $P(Y |X)$:

<center><img width = 500 src = "https://miro.medium.com/max/1400/1*l-3JCT9qwxWyLP_iVbzBmA.png"></img></center>

* The probability of a prediction will be high if it lies near the centre of the contour corresponding to its class and decreases as we move away from the centre of the contour.

**Example:**


we are interested in finding out a patient’s probability of having liver disease if they are an alcoholic.

* $A$  is the event “Patient has liver disease.” Past data tells us that 10% of patients have liver disease. $P(A) = 0.10$.
* $B$ is the test that “Patient is an alcoholic.” 5% of the patients are alcoholics. $P(B) = 0.05.$
* Among the patients diagnosed with liver disease, 7% are alcoholics. This is $B|A$: the probability that a patient is alcoholic, given that he has liver disease, is 7%.
* Bayes’ theorem tells us:
$P(A|B) = (0.07 * 0.1)/0.05 = 0.14$.
In other words, if the patient is an alcoholic, their chances of having liver disease is 0.14 (14%). This is a large increase from the 10% suggested by past data.

## Gaussian Discriminant Analysis [[ref](http://cs229.stanford.edu/notes2020spring/cs229-notes2.pdf)]

* Gaussian Discriminant Analysis(GDA) is a Generative Learning Algorithm.
* it tries to fit a Gaussian Distribution to every class of the data separately to capture the distribution of each class.
* Let’s consider a binary classification problem such that all the data samples are IID (Independently and Identically distributed).
* The probability density function for each individual class $x|y=0$  and $x|y=1$ are:
  * $x|y=0 \sim 𝑁(μ_0,∑)$
  * $x|y=1 \sim 𝑁(μ_1,∑)$

$μ_0$ and $μ_1$ are the means the two classes 0 and 1, respectively, of a dimention $R^n$. ∑ is a covariance matrix of a dimention $R^{n×n}$.
* As binary classification can either take value 1 or 0, the prior for each class can then be modeled as a Bernoulli distribution:
  * $y \sim Bernoulli(ϕ)$


* The join probability can be modeled  as **Likelihood function** as a function of the parameters of our chosen statistical models.

<center> $𝓛(ϕ,μ_0, μ_1, ∑|x,y)$</center>

* It equal to the product of probability distribution and class prior to each data sample.

<center> $𝓛(ϕ,μ_0, μ_1, ∑|x,y) = 𝚷_{i=0}^{m} p(x_i|y_i)p(y_i)$</center>

* we have to choose the value of parameters in a way to maximize the probability $𝓛$
* instead of maximizing the Likelihood function we can maximize Log-Likelihood $log(𝓛(ϕ,μ_0, μ_1, ∑|x,y))$
*  When maximizing th log-likelihood, the following paramter estimates are obtained.

<center> <img width=500 src="https://i.ibb.co/C9g3bLD/Capture-1.png" /></center>

Once we have estimated the optimal values of the parameters $ϕ$, $μ_0$, $μ_1$, and $∑$ using the maximum likelihood method, we can use these parameters to classify new data points based on their probability of belonging to each class.

For a new data point $x_{new}$, we can compute the probability of it belonging to class 0 and class 1 as follows:

* $p(y=0|x_{new}) = \frac{p(x_{new}|y=0)p(y=0)}{p(x_{new})}$
* $p(y=1|x_{new}) = \frac{p(x_{new}|y=1)p(y=1)}{p(x_{new})}$

$p(x_{new}|y=0)$ and $p(x_{new}|y=1)$ are the probability densities of the new data point $x_{new}$ given the class 0 and class 1, respectively, and can be computed using the Gaussian distributions with mean $μ_0$ and $μ_1$ and covariance matrix $∑$ that we estimated earlier.

$p(y=0)$ and $p(y=1)$ are the prior probabilities of class 0 and class 1, respectively, which can be modeled as a Bernoulli distribution with parameter $ϕ$.

$p(x_{new})$ is the marginal probability of $x_{new}$, which can be computed as: $p(x_{new}) = p(x_{new}|y=0)p(y=0) + p(x_{new}|y=1)p(y=1)$

Once we have computed the probabilities of $p(y=0|x_{new})$ and $p(y=1|x_{new})$, we can classify the new data point as follows:

* If $p(y=0|x_{new}) > p(y=1|x_{new})$, classify $x_{new}$ as class 0
* If $p(y=1|x_{new}) > p(y=0|x_{new})$, classify $x_{new}$ as class 1

## Naive Bayes Classifier [ref](http://www.cs.columbia.edu/~mcollins/em.pdf)

* It is a classification technique based on Bayes’ Theorem with an assumption of independence among predictors.
* Naive Bayes model is easy to build and particularly useful for very large data sets.
* Along with simplicity, Naive Bayes is known to outperform even highly sophisticated classification methods.
* Bayes theorem provides a way of calculating posterior probability $P(y|x)$ from $P(y)$, $P(x)$ and $P(x|y)$.

<center><img width = 300 src="https://i.ibb.co/YRrLG1z/Bayes-rule-300x172-300x172.png" /></center>

* for a set of features $X = \{x_1, x_2,...,x_n\}$, the Bayes theorem is:
<center>
$P(y| x_1, x_2,...,x_n) = \frac{P(x_1|y)P(x_2|y)...P(x_n|y)P(y)}{P(x_1)P(x_2)P...(x_n)}$
</center>

* The denominator can be ignored since it remains constant.

<center>
$P(y| x_1, x_2,...,x_n) ≃ P(y)\prod_{i=1}^{n} P (x_i|y)$
</center>

* The final step is to pick the hypothesis that is most probable ( maximuming posteriori)

<center>
${ {\hat {y}}={\underset {y\in Y}{\operatorname {argmax} }}\ p(y) \prod _{i=1}^{n}P(x_{i}\mid y).}$
</center>

* Maximizing the log-likelihood gives the following solutions:

* The method discussed above is applicable for discrete data.
* In case of continuous data, we need to make some assumptions regarding the distribution of values of each feature.
* The different naive Bayes classifiers differ mainly by the assumptions they make regarding the distribution of $P(x_i | y)$.


## Gaussian Naive Bayes classifier

* In Gaussian Naive Bayes, continuous values associated with each feature are assumed to be distributed according to a Gaussian distribution.
* The likelihood of the features is assumed to be Gaussian, hence, conditional probability is given by:

<center>${ p(x\mid y_{k})={\frac {1}{\sqrt {2\pi \sigma _{k}^{2}}}}\,e^{-{\frac {(x-\mu _{k})^{2}}{2\sigma _{k}^{2}}}}}$</center>

The following code shows how sklearn impliments Gaussian Naive Bayes classifier to estimate the distribution's paramaters values that best fit the data.

```python
from sklearn import cluster, datasets
import matplotlib.pyplot as plt
from sklearn.naive_bayes import GaussianNB
import numpy as np



blobs = datasets.make_blobs(n_samples=100, random_state=8)
X = blobs[0]
c = blobs[1]

gnb = GaussianNB()
gnb.fit(X, c)

# plot the dists
for mu, sigma in zip(gnb.theta_, np.sqrt(gnb.var_)):
  size = 100
  x = np.linspace(mu[0]-3*sigma[0], mu[0]+ 3*sigma[0], size)
  y = np.linspace(mu[1]-3*sigma[1], mu[1]+3*sigma[1], size)

  x, y = np.meshgrid(x, y)
  z = (1/(2*np.pi*sigma[0]*sigma[1]) * np.exp(-((x - mu[0])**2/(2*sigma[0]**2)
      + (y - mu[1]) **2/(2*sigma[1]**2))))
  plt.contourf(x, y, z, cmap='Blues')

plt.scatter(X[:, 0], X[:, 1],  c=c)

print('means :', gnb.theta_,'\n\n vars', gnb.var_)
```

## Decision trees [[ref]](https://pages.stat.wisc.edu/~loh/treeprogs/guide/wires11.pdf)

* Methods such as ***linear regression*** and ***Knn*** yield sets $A_j$ with piecewise linear/non-linear boundaries.
* These boundaries are not easy to interpret if the *number of features* is large.
* Classification tree methods yield rectangular sets $A_j$ by recursively partitioning the data set one $X$ variable at a time.
    


  **Algorithm**: Pseudocode for tree constructionby using exhaustive search
 1. Begins with the original set $S$ as the root node.
 2. iterates through the very unused attribute of the set $S$ and calculates Impurity($P$) and Information gain($IG$) of this attribute.
 3. Select the attribute $X^*$ which has the smallest Entropy or Largest Information gain.
 4. The set $S$ is then split by the selected attribute to produce a subset of the data.
 5. continue to recur on each subset until stop condition reached.

* An example tree which estimates the probability of ***kyphosis*** after spinal surgery, given the age of the patient and the vertebra at which surgery was started.
* Partitions (right) and decision tree structure (left) for a classification tree model with three classes labeled 1

## Node impurity and information gain:
*   The node impurity is a measure of the homogeneity of the labels at the node.

<center><img width=700 src ="https://i.ibb.co/3rZJGn0/Capture-3.png" /></center>

* Assuming that a split $s$ partitions the dataset $D$ of size $N$ into two datasets $Dleft$ and $Dright$ of sizes $Nleft$ and $Nright$, respectively, the information gain is:

$IG(D,s)=Impurity(D)− \frac{Nleft}{N}Impurity(Dleft)−\frac{Nright}{N}Impurity(Dright)$
* The split chosen at each tree node is chosen from the set argmaxsIG(D,s)

## Boosting

* Boosting is another ensemble technique to create a collection of predictors.
* In this technique, learners are learned sequentially .
* The procedure is to fit simple models to the data and then analyze data for errors.
* When an input is misclassified by a hypothesis, its weight is increased so that next hypothesis is more likely to classify it correctly.
* The main idea is to focus new learners on examples that others get wrong.
